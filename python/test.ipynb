```python
import os

def count_lines(directories):
    total_lines = 0
    file_count  = 0
    extensions  = {'.h', '.cpp'}

    print(f"{'File Path':<60} | {'Lines':>10}")
    print("-" * 75)

    excluded = ["stb_image.h", "TestECS.h", "stb_image_impl.cpp"]

    for directory in directories:
        for root, _, files in os.walk(directory):
            for file in files:
                if any(file.endswith(ext) for ext in extensions):

                    if file in excluded: continue

                    file_path = os.path.join(root, file)
                    
                    try:
                        with open(file_path, 'r', encoding='utf-8', errors='ignore') as f:
                            lines = sum(1 for line in f)
                            total_lines += lines
                            file_count += 1
                            # Print relative path for cleaner output
                            rel_path = os.path.relpath(file_path, directory)
                            print(f"{rel_path:<60} | {lines:>10}")
                    except Exception as e:
                        print(f"Could not read {file_path}: {e}")

    print("-" * 75)
    print(f"Total Files: {file_count}")
    print(f"Total Lines of Code: {total_lines}")

count_lines(["..\\include", "..\\src"])
```
